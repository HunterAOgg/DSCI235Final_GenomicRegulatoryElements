## Creation of Helper Functions and Core Datasets

Now, we will need a few helper functions before we can begin analysis. It may be necessary to limit our focus to a smaller sequence. This could be to equalize between different types of genes since all of the datasets have different sequence lengths included. This should also be able to normalize for the center, to ensure comparable genes such as promoters have conditions as similar as possible.

```python
def get_seq_crop(df,positions,center=0):
    ret_df = df.copy()
    positions = np.array(positions)+center
    ret_df['Sequence'] = [i[positions[0]:positions[1]+1] for i in ret_df['Sequence']]
    return ret_df
get_seq_crop(IRES_data,[-20,20],center=50).head()
```

The first function will calculate the frequency of a given sequence in a given dataframe. The second will calculate the frequency of the four canonical nucleic acids found in DNA. Examples of the usage of these functions can be seen below.

```python
def get_seq_freq(seq, arr):
    ret_freq = np.mean([i.count(seq)/len(i) for i in arr])
    return ret_freq

def get_acid_freq(arr):
    acids = ['A','T','C','G']
    ret_dict = {}
    for i in acids:
        ret_dict[i] = get_seq_freq(i,arr)
    return ret_dict

def calc_av_freq(seq_len):
    return (1/4) ** seq_len

def calc_adj_freq(seq,acid_freqs):
    ret_freq = np.array([acid_freqs[i] for i in seq]).prod()
    return ret_freq

print(get_seq_freq('C',IRES_data['Sequence']))
print(get_acid_freq(IRES_data['Sequence']))
print(get_acid_freq(splice_data['Sequence']))
```

The next set of functions will get the probabilities of a sequence appearing by chance. The first simply determines the frequency of a sequence assuming all nucleotides have an equal likelihood of appearing. However, not all nucleic acids are equally likely in sequence samples so the second accounts for this and adjusts accordingly. 

```python
def calc_av_freq(seq_len):
    return (1/4) ** seq_len

def calc_adj_freq(seq,acid_freqs):
    ret_freq = np.array([acid_freqs[i] for i in seq]).prod()
    return ret_freq

print(calc_av_freq(2))
print(calc_adj_freq('AC',get_acid_freq(IRES_data['Sequence'])))
```

While large sample sizes are generally desirable in data science, comparisons between groups can be unfairly weighted by those groups with larger sample sizes. The below function will return a dataframe with an equal number of samples for each group to allow for more accurate summary statistics. 

```python
def norm_n(df,label):
    min = df.groupby(label).count().min()[0]
    samples = df.groupby(label).sample(n=min)
    return samples
norm_n(full_df,'Type').head()
```

This means that they have a stronger () and regions rich in G and C are less likely to have . This second explanation is more likely as the major rate limiting step in transcription is currently believed to be promoter escape, the ability of the RNA polymerase to move past the intial dozen or so base pairs, in eukaryotes (with it present to a lesser degree in prokaryotes). 

## Consensus Sequences

Another important aim is to identify which positions are most well conserved for each element. Typically, the most essential elements (i.e. those which are directly recognized by macromolecules) are present with small variation in most genes which require them. These are known as 'consensus sequences'.

Mutations in these locations may cause disease and . It can also provide hints regarding the purpose of an individual sequence. For example, changes in promoter consensus sequences are known to be associated with changes in the level of transcription of a given gene.

```python
def get_cons_pos(arr,center,verbose=1):
    pos_to_acid = {0:'A',1:'T',2:'C',3:'G'}
    arr = np.array([np.array(list(i)) for i in arr],dtype=object)
    top_codons = [np.fromiter(get_acid_freq(i).values(),dtype=float) for i in arr.transpose()]
    top_codon_acids = [pos_to_acid[i.argmax()] for i in top_codons]
    top_codon_values = np.array([i.max() for i in top_codons])
    if verbose:
        ranked_norm = np.zeros(len(top_codon_values))
        sorted_norm = top_codon_values.copy()
        #sorted_norm.sort()
        ranked_norm = np.zeros(len(top_codon_values))#[np.where(ranked_norm==sorted_norm[i]) for i in range(64)]
        for i in range(len(top_codon_values)):
            max = sorted_norm.argmax()
            ranked_norm[i] = int(max)
            sorted_norm[max] = -1
        ranked_norm = ranked_norm-center
        return [top_codon_values,ranked_norm,top_codon_acids,top_codons]
    return top_codons_values
get_cons_pos(pro_promoter_data['Sequence'], center=pro_promoter_center,verbose=1)
```

These data show a distinctly well preserved group of codons centered around the -35th position, with the three most conserved positions being -34, -35, and -36. It is well known that the bacterial consensus sequence for the binding of sigma factors (proteins which are required for RNA polymerase to begin transcribing a gene and creating RNA8), and indeed this code replicates that finding. To confirm this, This is an excellent validation that implies it may be useful for finding other consensus sequences.

```python
top,ranked,acids,vals = get_cons_pos(pro_promoter_data['Sequence'], center=pro_promoter_center,verbose=1)
print(np.array(acids)[ranked.argsort()][pro_promoter_center-37:pro_promoter_center-31])
#print()
```

The consensus sequence produced in this data is 'GTAACA'. This is slightly different than the canonical -35 sequence of 'TTGACA', with only 4 of the 6 letters matching. Additionally, the results for the -10 consensus sequence (the Pribnow box) are present if weaker. Both the -10 and -11 positions are fairly well conserved.

This validation that implies this code may be useful for finding other consensus sequences, although it is far from a perfect system. 

```python
def plot_cons_pos(arr,center,title=False):
    top,ranked,acids,acid_values = get_cons_pos(arr, center=center,verbose=1)
    plt.figure(figsize=(8,8))
    #plt.ylim(bottom=-0.002,top=0.002)
    sorted_top = top[ranked.argsort()]
    #sorted_ranked = ranked.sort()
    ranked.sort()
    plt.plot(ranked, top, color='black')
    plt.xlabel('Position');
    plt.ylabel('Proportion of Sequences');
    if title:
        plt.title(title)
plot_cons_pos(pro_promoter_data['Sequence'],center=pro_promoter_center)
plot_cons_pos(splice_data['Sequence'],center=splice_center)
plot_cons_pos(human_promoter_data['Sequence'],center=human_promoter_center)
#plot

top,ranked,acids,freqs = get_cons_pos(splice_data['Sequence'],center=splice_center)
splice_seq = [acids[int(np.where(ranked == i)[0])] for i in range(-3,3)]
print(splice_seq)
```

## Appearance of Regulatory Sequences in Other Contexts

Now that these positions and sequences are identified, let us see how often these occur by chance in other places.

```python
#cons_seqs = {Splice: '', Euk_Pro: '', Pro_Pro: ''}
cons_seqs = ['TATAAT', 'TTGACA', 'TATAAA','CCTCCA','CCGCCC']

human_df = full_df[full_df['Organism'] == 'H. sapiens']
plot_seq_freq(human_df,'Type',['Coding','Promoter','Splice', 'IRES'],cons_seqs,)
for i in cons_seqs:
    for j in ['Coding','Promoter','Splice', 'IRES']:
        exp = calc_adj_freq(i,get_acid_freq(human_df[human_df['Type'] == j]['Sequence']))
        act = get_seq_freq(i,human_df[human_df['Type'] == j]['Sequence'])
        print(f'Expected frequency of {i} in {j} sequences is{exp: .5f}. Actual is{act: .5f}.')
```

Here, bacterial sequences ('TATAAT' and 'TTGACA' or the Pribnow box and -35 element) act as a control to ensure the calculations are properly made. These are expected to appear at approximately the rate of chance since they are not menaingful in eukaryotic genomes. These controls seem accurate enough, with the -35 element and the Pribnow box showed up at approximately the expected rates.

The TATAA box had higher than average prevalence in promoter sequences. This is expected in promoter sequences (since that is the canonical location). In the other three types of sequences, it was lower than average. This was especially notable in splicing sequences, where it was less than half the rate expected. This indicates that these sequencese are suboptimal evoluntarily, but still appear at a notable rate. 

Interestingly, the splicing consensus sequence is higher than average in all types of regulatory sequences observed. This is sensical for IRES, splicing, and coding sequences which all utilize splicing, but it is unclear why the actual is nearly twice the predicted frequency.

The human promoter consensus sequence is slightly elevated in all regulatory types except splicing sequences, but this is likely due to noise.
